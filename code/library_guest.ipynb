```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

```

```python
total_df = pd.read_excel('C:/workspace/이용자테마데이터_남녀대출분석_20210409030025.xlsx', sheet_name=0)

total_df["합계"]=(total_df["남성"]+total_df["여성"])

total_df.set_index("월", inplace=True)
total_df
```

```python
font_name = mpl.font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)

ax = total_df.plot(
    kind='bar', title='도서관 방문객 추이', figsize=(21, 9),
    legend=True, fontsize = 10, color=['#67D5B5', '#EE7785',
    '#84B1ED']
)
ax.set_xlabel('월', fontsize = 8)
ax.set_ylabel('남성/여성/총합', fontsize = 8)
ax.legend(['남성','여성','총합'], fontsize = 8)
```
