```python
import re
import sys
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
```

```python
def get_noun(news):
    
    okt=Okt()
    noun = okt.nouns(news)
#     for i,v in enumerate(noun):
#         if len(v) < 2:
#             noun.pop(i)
            
    count = Counter(noun)
    noun_list = count.most_common(100)
    
    return noun_list
```

```python
def visualize(noun_list):
    
    wc = WordCloud(font_path='Jalnan.ttf',\
              background_color = "white", \
              width = 1000, \
              height=1000, \
              max_words=100, \
              max_font_size=300)
    
    wc.generate_from_frequencies(dict(noun_list))
    wc.to_file('keyword_1word_contain_2020.png')
```

```python
def get_weighted_str(df):
    string = []

    for a in df.index:        
        for i in range(203-int(a)): # a=2 => range = 201
            ttl = df.loc[a]['도서명']
            sct = df.loc[a]['분야']
            string.append(ttl)
            string.append(sct)
    
    return ' '.join(string)
```

```python
if __name__=='__main__':
    filename = sys.argv[1]
    regex = "\(.*\)|\s-\s.*" # 집합연산자 데이터셋
    df = pd.read_excel('2020_best_seller.xls')
    
    for i in range(len(df)): # 괄호와 괄호 안 문자열 제거
        df['도서명'][i] = re.sub(regex, '', df['도서명'][i])    
    df.set_index('순위', inplace=True)

    words = get_weighted_str(df)
    noun_list = get_noun(words)
    visualize(noun_list)
```
