```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
%matplotlib inline

df_covid = pd.read_csv('C:/workspace/코로나_검색.csv', skiprows = [0, 1], encoding='euc-kr')
df_reading = pd.read_csv('C:/workspace/독서_검색.csv', skiprows = [0, 1], encoding='euc-kr')

#2열까지 스킵 후 불러오기
```

```python
df_covid = df_covid.rename({'코로나: (대한민국)':'코로나 검색추이'}, axis='columns')
df_reading = df_reading.rename({'독서: (대한민국)':'독서 검색추이'}, axis='columns')
```

```python
df_covid['코로나 검색추이'] = df_covid['코로나 검색추이'].replace('<1', 1)
df_reading['독서 검색추이'] = df_reading['독서 검색추이'].replace('<1', 1)
#replace 로 0<1 값를 1로 변경
```

```python
#주로 내림차순
covid = df_covid.sort_values( by = "주", ascending = True)
reading = df_reading.sort_values( by = "주", ascending = True)
covid, reading
```

```python
covid.set_index("주", inplace=True)
reading.set_index("주", inplace=True)

covid, reading
```

```python
covid = covid['코로나 검색추이'].astype(float) #검색량 항목 모두 숫자열로 변경 #변환안할시 오류
covid.plot()
```

```python
reading = reading['독서 검색추이'].astype(float) 
reading.plot()
```

```python
covid_reading = pd.merge(covid, reading, left_index=True, right_index=True, how='right')
covid_reading
```

```python
covid_reading.plot()
```

```python
font_name = mpl.font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
covid_reading.plot(kind = 'bar', title='google 코로나&독서 검색비교', figsize=(21, 9), legend=True, fontsize = 11, color=['#67D5B5', '#EE7785'], xlabel = '2019 - 2021 주간 데이터')
```
